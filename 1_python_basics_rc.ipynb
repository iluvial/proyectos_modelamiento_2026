Ejercicio 1

```python
precip_months=['jan','feb','mar','apr','may','jun','jul','aug','sept','oct','nov','dec']
print(precip_months)
precip_inches=[0.7,0.75,1.85,2.93,3.05,2.02,1.93,1.62,1.84,1.31,1.39,0.84]
print(precip_inches)
```

Ejercicio 2

```python
precip_mm = [inch * 25.4 for inch in precip_inches]
print(precip_mm)
```

Ejercicio 3

```python
all_precip_data = [precip_months,precip_mm]
print(all_precip_data)
```

Ejercicio 4

```python
# Importar la librería necesaria
import matplotlib.pyplot as plt

# Crear figura y ejes
fig, ax = plt.subplots(figsize=(6, 6))

# Graficar datos (modifica estas líneas)
ax.bar(
    all_precip_data[0],
    all_precip_data[1],
    color="aqua"
)

# Configurar títulos y etiquetas
ax.set(
    title="Precitacion Promedio Mensual",
    xlabel="Meses del año",
    ylabel="Precipitacion promedio(mm)"
)

plt.show()
```
